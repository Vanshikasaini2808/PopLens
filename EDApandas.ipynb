```python
# Select only numeric columns for correlation and store in a variable
numeric_df = df.select_dtypes(include=['number'])

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Heatmap of Correlation")
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['2020 Population'], y=df['2010 Population'])
plt.xlabel("2020 Population")
plt.ylabel("2010 Population")
plt.title("Scatter Plot of Population in 2020 vs 2010")
plt.show()
```

```python
# Reshaping the dataset to have Year and Population columns
year_columns = ['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2020 Population']
df_melted = df.melt(id_vars=['Country'], value_vars=year_columns, var_name='Year', value_name='Population')
```

```python
# Convert 'Year' column from strings like '1970 Population' to just year numbers
df_melted['Year'] = df_melted['Year'].astype(str).str.extract(r'(\d+)').astype(int)
```

```python
# Line Graph
plt.figure(figsize=(10, 6))
sns.lineplot(x=df_melted['Year'], y=df_melted['Population'], marker='o')
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Growth Over Time")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x='Year', y='Population', data=df_melted)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Box Plot of Population over Years')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
